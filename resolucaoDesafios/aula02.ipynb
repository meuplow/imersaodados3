###Desafio 07 - Deixar os gráficos bonitões (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt
```

```python
valor_tempo = dados['tempo'].value_counts(ascending=True)
valor_tempo.sort_index()
```

```python
plt.figure(figsize=(15, 10))
valor_tempo = dados['tempo'].value_counts(ascending=True)
ax = valor_tempo.sort_index().plot.bar()
ax.set_title('Janelas de tempo', fontsize=25)
ax.set_xlabel('Tempo', fontsize=20)
ax.set_ylabel('Frequência', fontsize=20)
plt.xticks(rotation = 0, fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

###Desafio 01 - Ordenar o gráfico countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(6, 4))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)

ax.set_title('Top 5 compostos')
ax.set_ylabel('Frequência')
ax.set_xlabel('Compostos')
plt.show()
```

###Desafio 02 - Melhorar a visualização alterando tamanho da fonte

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos, palette='Paired')

ax.set_title('Top 5 compostos', size=25)
ax.set_ylabel('Frequência', size=18)
ax.set_xlabel('Nome dos compostos', size=18)
plt.show()
```
