###Desafio 3 - Proporção das classes tratamento

```python
#retorna os dados normalizados
dadosNormalizados = dados['tratamento'].value_counts(normalize = True)

#transforma os valores em percentual
percentual = dadosNormalizados * 100 

#exibe os percentuais com duas casas decimais
round(percentual,2)
```

###Desafio 4 - Quantas tipos de drogas foram investigados

###Desafio 6 - Renomear as colunas tirando o hífen


```python
# Retira todos os hifens dos titulos das colunas
dados.columns = [coluna.replace('-', '_') for coluna in dados.columns]
dados.head()
```

###Desafio 5 - Procurar na documentação o método query(pandas)

```python
#consulta e retorna os dados filtrados
dados.query('g_0 > 0').head()
```

###Desafio 07 - Deixar os gráficos bonitões (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt
```

```python
valor_tempo = dados['tempo'].value_counts(ascending=True)
valor_tempo.sort_index()
```

```python
plt.figure(figsize=(15, 10))
valor_tempo = dados['tempo'].value_counts(ascending=True)
ax = valor_tempo.sort_index().plot.bar()
ax.set_title('Janelas de tempo', fontsize=25)
ax.set_xlabel('Tempo', fontsize=20)
ax.set_ylabel('Frequência', fontsize=20)
plt.xticks(rotation = 0, fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
